# Connect to Database

```python
# Membuka koneksi ke database SQLite (akan dibuat jika tidak ada)
conn = sqlite3.connect('logistic.db')
cursor = conn.cursor()

#import os
#os.remove('logistic.db')
```

# Build a Table

```python
# Membuat tabel customer
cursor.execute('''
CREATE TABLE IF NOT EXISTS customer (
    C_ID INTEGER PRIMARY KEY,
    C_NAME TEXT(30),
    C_EMAIL_ID TEXT(50),
    C_CONT_NO INTEGER,
    C_ADDR TEXT(100),
    C_TYPE TEXT(30),
    M_ID INTEGER,
    FOREIGN KEY (M_ID) REFERENCES membership(M_ID)
)
''')

# Membuat tabel membership
cursor.execute('''
CREATE TABLE IF NOT EXISTS membership (
    M_ID INTEGER PRIMARY KEY,
    Start_date TEXT,
    End_date TEXT
)
''')

# Membuat tabel payment_details
cursor.execute('''
CREATE TABLE IF NOT EXISTS payment_details (
    Payment_ID TEXT(40) PRIMARY KEY,
    AMOUNT INTEGER,
    Payment_Status TEXT(10),
    Payment_Date TEXT,
    Payment_Mode bn TEXT(25),
    SH_ID INTEGER,
    C_ID INTEGER,
    FOREIGN KEY (SH_ID) REFERENCES shipment_details(SH_ID)
    FOREIGN KEY (C_ID) REFERENCES customer(C_ID)
)
''')

# Membuat tabel status
cursor.execute('''
CREATE TABLE IF NOT EXISTS status (
    Current_Status TEXT(15),
    Sent_date TEXT,
    Delivery_date TEXT,
    SH_ID INTEGER PRIMARY KEY
    
)
''')

# Membuat tabel shipment_details
cursor.execute('''
CREATE TABLE IF NOT EXISTS shipment_details (
    SH_ID INTEGER PRIMARY KEY,
    SH_CONTENT TEXT(40),
    SH_DOMAIN TEXT(15),
    SER_TYPE TEXT(15),
    SH_WEIGHT TEXT(10),
    SH_CHARGES TEXT(10),
    SR_ADDR TEXT(100),
    DS_ADDR TEXT(100),
    C_ID ,
    FOREIGN KEY (C_ID) REFERENCES customer(C_ID)
)
''')

# Membuat tabel employee manages shipment
cursor.execute('''
CREATE TABLE IF NOT EXISTS employee_manages_shipment (
    Employee_E_ID INTEGER,
    Shipment_Sh_ID INTEGER,
    Status_Sh_ID INTEGER,
    FOREIGN KEY (Employee_E_ID) REFERENCES employee_details(E_ID),
    FOREIGN KEY (Shipment_Sh_ID) REFERENCES shipment_details(SH_ID),
    FOREIGN KEY (Status_Sh_ID) REFERENCES status(SH_ID)
)
''')

# Membuat tabel employee details
cursor.execute('''
CREATE TABLE IF NOT EXISTS employee_details (
    E_ID INTEGER PRIMARY KEY,
    E_NAME TEXT(30),
    E_BRANCH TEXT(15),
    E_DESIGNATION TEXT(40),
    E_ADDR TEXT(100),
    E_CONT_NO INTEGER
)
''')

# Commit perubahan untuk memastikan tabel dibuat
conn.commit()
```

# Fill the table

```python
# Menambahkan data dari customer.csv ke dalam tabel customer
for index, row in customer_df.iterrows():
    cursor.execute('''
    INSERT INTO customer (C_ID, C_NAME, C_EMAIL_ID, C_CONT_NO, C_ADDR, C_TYPE, M_ID)
    VALUES (?, ?, ?, ?, ?, ?, ?)
    ''', (row['C_ID'], row['C_NAME'], row['C_EMAIL_ID'], row['C_CONT_NO'], row['C_ADDR'], row['C_TYPE'], row['M_ID']))

# Menambahkan data dari membership.csv ke dalam tabel membership
for index, row in membership_df.iterrows():
    cursor.execute('''
    INSERT INTO membership (M_ID, Start_date, End_date)
    VALUES (?, ?, ?)
    ''', (row['M_ID'], row['Start_date'], row['End_date']))

# Menambahkan data dari payment_details.csv ke dalam tabel payment_details
for index, row in payment_details_df.iterrows():
    cursor.execute('''
    INSERT INTO payment_details (Payment_ID, AMOUNT, Payment_Status, Payment_Date, Payment_Mode, SH_ID, C_ID)
    VALUES (?, ?, ?, ?, ?, ?, ?)
    ''', (row['Payment_ID'], row['AMOUNT'], row['Payment_Status'], row['Payment_Date'], row['Payment_Mode'], row['SH_ID'], row['C_ID']))

# Menambahkan data dari status.csv ke dalam tabel status
for index, row in status_df.iterrows():
    cursor.execute('''
    INSERT INTO status (Current_Status, Sent_date, Delivery_date, SH_ID)
    VALUES (?, ?, ?, ?)
    ''', (row['Current_Status'], row['Sent_date'], row['Delivery_date'], row['SH_ID']))

# Menambahkan data dari shipment_details.csv ke dalam tabel shipment_details
for index, row in shipment_details_df.iterrows():
    cursor.execute('''
    INSERT INTO shipment_details (SH_ID, SH_CONTENT, SH_DOMAIN, SER_TYPE, SH_WEIGHT, SH_CHARGES, SR_ADDR, DS_ADDR, C_ID)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (row['SH_ID'], row['SH_CONTENT'], row['SH_DOMAIN'], row['SER_TYPE'], row['SH_WEIGHT'], row['SH_CHARGES'], row['SR_ADDR'], row['DS_ADDR'], row['C_ID']))

# Menambahkan data dari employee_manages_shipment.csv ke dalam tabel employee_manages_shipment
for index, row in employee_manages_shipment_df.iterrows():
    cursor.execute('''
    INSERT INTO employee_manages_shipment (Employee_E_ID, Shipment_Sh_ID, Status_Sh_ID)
    VALUES (?, ?, ?)
    ''', (row['Employee_E_ID'], row['Shipment_Sh_ID'], row['Status_Sh_ID']))

# Menambahkan data dari employee_details.csv ke dalam tabel employee_details
for index, row in employee_details_df.iterrows():
    cursor.execute('''
    INSERT INTO employee_details (E_ID, E_NAME, E_BRANCH, E_DESIGNATION, E_ADDR, E_CONT_NO)
    VALUES (?, ?, ?, ?, ?, ?)
    ''', (row['E_ID'], row['E_NAME'], row['E_BRANCH'], row['E_DESIGNATION'], row['E_ADDR'], row['E_CONT_NO']))

# Commit perubahan untuk memastikan data dimasukkan
conn.commit()
```

# Check Relationship Table

```python
# Menampilkan data customer yang terhubung dengan membership dan payment_details
cursor.execute('''
SELECT c.C_NAME, c.C_EMAIL_ID, m.Start_date, m.End_date, p.AMOUNT, p.Payment_Status
FROM customer c
JOIN membership m ON c.M_ID = m.M_ID  -- Menghubungkan customer dengan membership
JOIN payment_details p ON c.C_ID = p.C_ID  -- Menghubungkan customer dengan payment_details
''')

# Menampilkan hasil query
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## 8. Finding C_ID, M_ID, and tenure for those customers whose membership is over 10 years.

```python
# Menampilkan C_ID, M_ID, dan tenure untuk pelanggan dengan masa keanggotaan lebih dari 10 tahun
cursor.execute('''
SELECT c.C_ID, m.M_ID, 
       (JULIANDAY('now') - JULIANDAY(m.Start_date)) / 365 AS Tenure
FROM customer c
JOIN membership m ON c.M_ID = m.M_ID
WHERE (JULIANDAY('now') - JULIANDAY(m.Start_date)) / 365 > 10
''')

# Menampilkan hasil query
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## 9. Considering average payment amount based on customer type having payment mode as COD in descending order. 

```python
# Menampilkan rata-rata jumlah pembayaran berdasarkan tipe pelanggan (C_TYPE) dengan mode pembayaran COD dan mengurutkan berdasarkan jumlah pembayaran tertinggi
cursor.execute('''
SELECT c.C_TYPE, AVG(p.AMOUNT) AS Average_Payment_Amount
FROM customer c
JOIN payment_details p ON c.C_ID = p.C_ID
WHERE p.Payment_Mode = 'COD'
GROUP BY c.C_TYPE
ORDER BY Average_Payment_Amount DESC
''')

# Menampilkan hasil query
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## 10. Calculate the average payment amount based on payment mode where the payment date is not null. 

```python
# Menampilkan rata-rata jumlah pembayaran berdasarkan mode pembayaran dimana tanggal pembayaran tidak null
cursor.execute('''
SELECT p.Payment_Mode, AVG(p.AMOUNT) AS Average_Payment_Amount
FROM payment_details p
WHERE p.Payment_Date IS NOT NULL
GROUP BY p.Payment_Mode
''')

# Menampilkan hasil query
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## 11. Calculate the average shipment weight based on payment_status where shipment content does not start with "H."

```python
# Menampilkan rata-rata berat pengiriman berdasarkan status pembayaran dimana konten pengiriman tidak dimulai dengan "H"
cursor.execute('''
SELECT p.Payment_Status, AVG(s.SH_WEIGHT) AS Average_Shipment_Weight
FROM payment_details p
JOIN shipment_details s ON p.SH_ID = s.SH_ID
WHERE s.SH_CONTENT NOT LIKE 'H%' 
GROUP BY p.Payment_Status
''')

# Menampilkan hasil query
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## 14.  Find the membership start and end dates for customers with 'Paid' payment status.

```python
# Menampilkan start date dan end date membership untuk pelanggan dengan status pembayaran 'Paid'
cursor.execute('''
SELECT m.Start_date, m.End_date
FROM payment_details p
JOIN customer c ON p.C_ID = c.C_ID
JOIN membership m ON c.M_ID = m.M_ID
WHERE p.Payment_Status = 'PAID'
''')

# Menampilkan hasil query
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## 15. List the clients who have made 'Card Payment' and have a 'Regular' service type.

```python
# Menampilkan pelanggan yang melakukan pembayaran 'Card Payment' dan memiliki jenis layanan 'Regular'
cursor.execute('''
SELECT c.C_NAME, c.C_EMAIL_ID, p.Payment_Mode, s.SH_DOMAIN
FROM payment_details p
JOIN customer c ON p.C_ID = c.C_ID
JOIN shipment_details s ON c.C_ID = s.C_ID
WHERE p.Payment_Mode = 'CARD PAYMENT' 
AND s.SER_TYPE = 'Regular'
''')

# Menampilkan hasil query
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## 16. Calculate the average shipment weight for each shipment domain (International and Domestic).

```python
# Menampilkan rata-rata berat pengiriman berdasarkan domain pengiriman (International dan Domestic)
cursor.execute('''
SELECT SH_DOMAIN, AVG(CAST(SH_WEIGHT AS REAL)) AS average_weight
FROM shipment_details
WHERE SH_DOMAIN IN ('International', 'Domestic')
GROUP BY SH_DOMAIN
''')

# Menampilkan hasil query
rows = cursor.fetchall()
for row in rows:
    print(row)
```

## 17. Identify the shipment with the highest charges and the corresponding client's name

```python
# Menampilkan pengiriman dengan biaya tertinggi dan nama klien yang bersangkutan
cursor.execute('''
SELECT s.SH_ID, s.SH_CHARGES, c.C_NAME
FROM shipment_details s
JOIN customer c ON s.C_ID = c.C_ID
ORDER BY s.SH_CHARGES DESC
LIMIT 1
''')

# Menampilkan hasil query
row = cursor.fetchone()
if row:
    print(f"Shipment ID: {row[0]}, Charges: {row[1]}, Client Name: {row[2]}")
else:
    print("No data found.")
```

## 18.  Count the number of shipments with the 'Express' service type that are yet to be delivered.

```python
# Menampilkan jumlah pengiriman dengan 'Express' service type yang belum terkirim (belum ada delivery date)
cursor.execute('''
SELECT COUNT(*)
FROM shipment_details s
JOIN status st ON s.SH_ID = st.SH_ID
WHERE s.SER_TYPE = 'Express'
AND st.DELIVERY_DATE IS NULL
''')

# Menampilkan hasil query
row = cursor.fetchone()
if row:
    print(f"Number of 'Express' shipments yet to be delivered: {row[0]}")
else:
    print("No data found.")
```
